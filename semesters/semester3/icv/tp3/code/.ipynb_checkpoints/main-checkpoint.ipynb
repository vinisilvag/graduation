## __Funções auxiliares__

Aqui serão definidas algumas funções auxiliares que serão utilizadas ao longo do código.

* show()
    > Recebe uma imagem e a legenda dela e plota essa imagem no bloco de código do _notebook_.

* segment()
    > Recebe a imagem de entrada (_captcha_) e o tamanho do padding a ser aplicado de cada lado e realiza esse padding, preenchendo de branco os casos onde a imagem ficaria menor do que 50x50 e mantendo o _overlapping_ entre as letras, caso esse exista. O _overlapping_ foi mantido para que, quando cada segmento for enviado para a rede, não falte nenhuma parte do caractere analisado (após alguns testes, concluímos que era melhor que o outro caractere aparecesse parcialmente, em detrimento ao caractere desejado sair cortado na segmentação).
    
* file_name()
    > Recebe um diretório e define qual será o nome (numérico) de um arquivo, baseado em quantos arquivos existem previamente naquela pasta.
    
* get_label()
    > Recebe um diretório como parâmetro, lê uma label desse diretório e a retorna como tipo __string__.

````python
def show(img, caption="Imagem"):
    # Carregando imagem
    plt.clf()
    plt.figure()
    plt.title(caption)
    plt.imshow(img, cmap="gray")
    plt.show()
    plt.close()
    
def segment(img, border = 10):
    segs = []
    cols = img.shape[1]
    prev = 0
    for i in range(int(cols/6), cols+1, int(cols/6)):
        pad  = 0
        
        # Define if padding should happen
        if prev - border < 0:
            left = 0
            pad  = prev - border
        else:
            left = prev-border
        
        if i + border > cols:
            right = cols
            pad   = i + border - cols
        else:
            right = i + border
        
        # Segment
        seg = img[:,left:right]
        
        # Padding
        if pad < 0:
            seg = np.append(np.full((seg.shape[0], -pad), 255), seg, axis = 1)
        elif pad > 0:
            seg = np.append(seg, np.full((seg.shape[0], pad), 255), axis = 1)
        segs.append(seg)
        prev = i
    
    return segs

def file_name(dir):
    n_files = len([1 for x in list(os.scandir(dir)) if x.is_file()])
    return str(n_files+1).zfill(4)

def get_label(path):
    with open(path) as f:
        return f.read().strip()
````

## __Pré-processamento__

Para treinar o classificador, optamos pela estratégia de treiná-lo separadamente, para identificar cada caractere possível (0-9 e A-Z). No entanto, para facilitar esse treinamento, pré-processamos os dados do conjunto de __treinamento__ e do conjunto de __validação__ separadamente e depois definimos/treinamos o classificador. A estratégia de pré-processamento foi a seguinte: cada imagem presente nos conjuntos __treinamento__, __validação__ e __teste__ será segmentada nos seus 6 caracteres. Cada caractere segmentado vai ser comparado com seu ground-truth e redirecionado para uma pasta separada, que contém somente caracteres da mesma classe. 

Como o nome da pasta é igual ao _ground-truth_ para aquele segmento, conseguimos organizar todos os dados em suas respectivas classes e, com isso, poderemos usar alguns utilitários do Keras para facilitar o treinamento do classficador, a identificação das classes presentes no _dataset_, etc; auxiliando tanto semânticamente, para entendimento do _dataset_ como um todo quanto o código criado para importar o _dataset_ tratado e manipulá-lo.

Além disso, durante esse processo, notamos que alguns _captchas_ continham caracteres não reconhecidos (?) e/ou 7 caracteres no seu ground-truth. Para evitar que esses _captchas_ causassem inconsistências na hora do treinamento/validação e/ou teste, eles foram removidos durante o pré-processamento dos captchas. Sendo assim, nenhum desses captchas foi considerado, tanto no treinamento quanto no teste do classificador.

Após o final do pré-processamento, uma pasta chamada __gen_data__ terá sido criada e terá a seguinte estrutura:

```
project
└─── code
│   │   tp3-icv.ipynb
│   
└─── data
│    └─── ...
│   
└─── gen_data
     └─── teste
     │    └─── 0
     │    │    │ 0001.jpg
     │    │    │ 0002.jpg
     │    │    │ 0003.jpg
     │    │    │ ...
     │    │
     │    └─── 1
     │    └─── 2
     │    └─── 3
     │    └─── ...
     │
     └─── treinamento
     │    └─── ...
     │
     └─── validacao
          └─── ...
```

````python
def pre_process(img, label, folder, name):
    # Defining general folder
    directory = os.path.join("../gen_data", folder)
    if not os.path.exists(directory):
        os.makedirs(directory)
    
    # Segmenting image
    segs = segment(img)
    if(len(segs) != len(label)):
        print("Error! Got size of segs:", len(segs), "But size of label:", len(label), "on file", name)
        return
        
    for i in range(len(label)):
        y    = label[i]
        if y == '?':
            continue
            
        # Defining label folder
        path = os.path.join(directory, y)
        if not os.path.exists(path):
            os.makedirs(path)
        
        # Saving image
        filename = file_name(path) + ".png"
        filename = os.path.join(path, filename)
        
        cv2.imwrite(filename, segs[i])
````

````python
folders = ["treinamento", "teste", "validacao"]
root_path  = "../data/CAPTCHA-10k"
label_root = "../data/CAPTCHA-10k/labels10k"

for folder in folders:
    directory = os.path.join(root_path, folder)
    for filename in tqdm(os.listdir(directory)):
        f       = os.path.join(directory, filename)
        if not os.path.isfile(f):
            continue
        label_f = os.path.join(label_root, filename[:-3] + "txt")
        
        # Reading image and label
        src_img = cv2.imread(f, 0)
        label   = get_label(label_f)
        
        pre_process(src_img, label, folder, filename)
````

## __Parâmetros pré-modelo__

Nesta seção, alguns parâmetros gerais serão definidos, como os parâmetros dos segmentos, o diretório e a importação dos _datasets_ após processamento, etc.

Utilizamos a função _image_dataset_from_directory()_ do Keras para importar os _datasets_ de treinamento e validação pós-processamento e, após isso, verificamos os resultados, visualizando primeiramente quais foram as classes identificadas pelo Keras. Além disso, alguns dos dados importados do conjunto de treinamento, apenas para verificar como ficaram os segmentos e para saber se a importação ocorreu corretamente.

````python
batch_size = 32
img_height = 50
img_width  = 50

train_dir = "../gen_data/treinamento/"
val_dir   = "../gen_data/validacao/"
````

````python
# Training Dataset
train_ds = tf.keras.utils.image_dataset_from_directory(
  train_dir,
  seed=123,
  image_size=(img_height, img_width),
  color_mode="grayscale",
  batch_size=batch_size)

# Validation Dataset
val_ds = tf.keras.utils.image_dataset_from_directory(
  val_dir,
  seed=123,
  image_size=(img_height, img_width),
  color_mode="grayscale",
  batch_size=batch_size)
````

Classes Definidas:

````python
class_names = train_ds.class_names
print(class_names)
````

Exemplos das imagens segmentadas:

````python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 10))
for images, labels in train_ds.take(1):
    for i in range(9):
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"), cmap="gray")
        plt.title(class_names[labels[i]])
        plt.axis("off")
````

## __Arquitetura da Rede__

Aqui são definidas todas as camadas da rede neural que será montada e treinada.

Muitas tentativas foram feitas, realizando mudanças na quantidade de camadas convolucionais, na quantidade de filtros em casa camada, na quantidade de parâmetros treináveis da rede ___fully connected___, etc. Ao final, chegamos a conclusão que uma rede neural profunda, com 3 camadas convolucionais se uma rede __fully connected__ com apenas 128 parâmetros foi suficiente para que obtivéssemos um resultado extremamente interessante e tivéssemos "poucos" parâmetros treináveis e uma rede com treinamento relativamente rápido.

Adicionamos camadas de Max Pooling após cada camada convolucional, para que o tamanho dos _feature maps_ fosse gradativamente reduzido e o processo de convolução tomasse menos processamento e menos pesos devessem ser identificados. Além disso, acrescentamos uma camada de ___rescaling___ antes das camadas convolucionais, visto que a CNN tende a apresentar melhores resultados (mais consistentes) se os valores da imagem estiverem normalizados. Por fim, uma camada de ___dropout___ foi adicionado antes do _flatten_ da última camada, para que 20% dos pesos fossem descartados e pudéssemos evitar o ___overfitting___ no conjunto de treinamento.

Observamos que aumentar o número de filtros nas camadas convolucionais de maneira gradativa trazia bons resultados, por isso 16, 32 e 64 como a quantidade de filtros para cada camada convolucional. 

Ao final, o classificador consiste de apenas duas camadas ___fully connected___, que são responsáveis por relacionar as features obtidas e classificar cada segmento, a partir dessas features.

````python
num_classes = len(class_names)

model = Sequential([
    layers.Rescaling(1./255, input_shape=(img_height, img_width, 1)),
    layers.Conv2D(16, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(32, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(64, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    layers.Dropout(0.2),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(num_classes, activation='softmax')
])
````

````python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
````

__Sumário do Modelo:__

__Obs__: Foi corrigido o erro apontado em sala de aula que fazia com que a rede analisasse as imagens com três canais. As imagens foram passadas para grayscale e não foram observadas diferenças relevantes na performance do modelo.

````python
model.summary()
````

### __Treinamento__

4 épocas. Observamos que para mais épocas, o modelo começa a ter _overfitting_ e os resultados já não eram mais tão significantes. Por isso, optamos por manter apenas 4 épocas.

Batch Size 32.

O tempo de treinamento varia de 30 à 120 segundos por época dependendo das configurações da máquina em que for executado.

````python
epochs = 4

history = model.fit(
  train_ds,
  validation_data=val_ds,
  epochs=epochs
)
````

## __Resultados do treinamento/validação__

Aqui exibimos os resultados da rede após o treinamento e validação.

É possível perceber que a acurácia final do treino não destoa muito da de validação, um bom sinal de que a rede não está sofrendo com Overfitting.

````python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(epochs)

plt.figure(figsize=(8, 8))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
````

## __Teste por Caractere__

Iremos construir a matriz de confusão para cada caractere, assim podemos analisar métricas como Precisão e Revocação para cada caractere separadamente.

Para tal, iremos passar todas as imagens do dataset de teste de cada caractere pela rede e analisar se a resposta é igual ou diferente da resposta esperada.

````python
TP    = {c:0 for c in class_names}
FP    = {c:0 for c in class_names}
TT    = {c:0 for c in class_names}

total = 0

hist  = {c:{c:0 for c in class_names} for c in class_names}
for c in class_names:
    print("Class", c)
    directory = "../gen_data/teste/"+c+"/"
    for filename in tqdm(os.listdir(directory)):
        img_path = os.path.join(directory, filename)
        if not os.path.isfile(img_path):
            continue
        img = tf.keras.utils.load_img(
            img_path, target_size=(img_height, img_width), color_mode="grayscale",
        )
        img_array = tf.keras.utils.img_to_array(img)
        img_array = tf.expand_dims(img_array, 0) # Create a batch

        predictions = model.predict(img_array, verbose = 0)
        score       = tf.nn.softmax(predictions[0])
        pc          = class_names[np.argmax(score)]
        
        if pc == c:
            TP[c] += 1
        else:
            FP[pc] += 1
        
        hist[c][pc] += 1
        
        TT[c] += 1
        total += 1
        
FN = {c: TT[c] - TP[c] for c in class_names}

precision = {c: TP[c] / (TP[c] + FP[c]) for c in class_names}
recall    = {c: TP[c] / (TP[c] + FN[c]) for c in class_names}
````

Com a matriz de confusão e as métricas calculadas para cada caractere, podemos fazer as médias para definir uma Precisão, Revocação e Acurácia média para a rede. 

````python
med_prec = np.mean([x for x in precision.values()])
med_rec  = np.mean([x for x in recall.values()])
med_acc  = (total - np.sum([FP[c] + FN[c] for c in class_names])) / total
````

Precisão, Revocação e Acurácia Médias:

````python
print("Precisão:", med_prec)
print("Recall:",   med_rec)
print("Acurácia:", med_acc)
````

Agora utilizaremos a biblioteca Pandas para gerar um Dataframe com as métricas para cada caractere.

````python
df1 = pd.DataFrame([precision]).T
df2 = pd.DataFrame([recall]).T
df  = pd.concat([df1,df2], axis=1)
df  = df.set_axis(['Precision', 'Recall'], axis=1, inplace=False)
````

## ___Pipeline_ de reconhecimento__

Aqui o _pipeline_ de reconhecimento foi definido.

Primeiramente, os labels de teste e o _dataset_ de teste são importados. Após isso, cada _captcha_ do _dataset_ será percorrida e segmentada. A CNN vai predizer qual a classe de cada segmento. Ao final, os resultados das predições são concatenados em uma __string__ e comparados com o _ground-truth_ para aquele _captcha_. Juntamente com isso, calculamos algumas métricas, como a quantidade de acertos por número de caracteres (rede acertou um caractere, dois caracteres, três caracteres e assim por diante). Repetimos esse processo para todas as 1000 imagens ao final exibimos esses resultados estatísticos, para que análises sobre a qualidade da rede possam ser feitos.

````python
labels_files = sorted(glob.glob('../data/CAPTCHA-10k/labels10k/*.txt'))
labels = [open(file, "r").read(6) for file in labels_files]
test_labels  = labels[9000:10000]

test_dataset_filelist = sorted(glob.glob('../data/CAPTCHA-10k/teste/*.jpg'))
test_dataset  = np.array([np.array(Image.open(file).getchannel('R')) for file in test_dataset_filelist])
````

````python
# A exibição das imagens foi comentada para que esse bloco de código execute mais rápido

samples_number = 1000

predict = []
matches = np.zeros(6)

for idx, image in tqdm(enumerate(test_dataset[:samples_number,:])):
    # show(image)
    
    segs = segment(image)
    
    captcha = []
    
    for seg in segs:
        # show(seg, "Segmento")
        pre_processed = tf.keras.utils.img_to_array(seg)
        pre_processed = tf.expand_dims(pre_processed, 0)
    
        predictions = model.predict(pre_processed, verbose=0)
        score = predictions[0][np.argmax(predictions)]
    
        # print("Classe Predita:", class_names[np.argmax(predictions)])
        # print("Confiança:", 100 * np.max(score))
        # print("\n")
            
        captcha.append(class_names[np.argmax(predictions[0])])
        
    stringified = ''.join(captcha)
    
    match = 0
    
    for i in range(6):
        if stringified[i] == test_labels[idx][i]:
            match += 1
            
    for i in range(match):
        matches[i] += 1
                
    predict.append(stringified)
    
    print("Resultado esperado:", test_labels[idx])
    print("Resultado do classificador:", stringified)
    
    print("\n")
    
matches = matches / samples_number
````

Como pode ser observado nos resultados, para o reconhecimento do _captcha_ completo, conseguimos aproximadamente 82% de acurácia, que é um resultado bastante interessante, dado a simplicidade da rede. Além disso, obtivemos uma precisão e um _recall_ de aproximadamente 70% e ótimos resultados para o reconhecimento dos caracteres individualmente, contando com praticamente 100% de reconhecimento para pelo menos 4 caracteres, 98% para 5 e 82% para 6.

````python
plt.plot([1, 2, 3, 4, 5, 6], matches, label='Classification Accuracy')
plt.legend(loc='lower left')
plt.title('Character Recognition Rate')
plt.show()

predict = np.array(predict)
ground_truth = np.array(test_labels[:samples_number])

accuracy = accuracy_score(ground_truth, predict)
precision, recall, _, _ = precision_recall_fscore_support(ground_truth, predict, average='macro', zero_division=0)

for idx, match in enumerate(matches):
    print("{} acerto(s): {:2.2%}".format(idx+1, match))

print("\nAcurácia:", accuracy)
print("Precisão:", precision)
print("Recall:", recall)
````
