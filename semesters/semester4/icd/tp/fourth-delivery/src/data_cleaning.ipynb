```python
import pandas as pd
```

```python
# Importando e exibindo as primeiras linhas do dataset

tiktok = pd.read_csv('../data/tiktok.csv')
tiktok.head()
```

```python
print("Número de registros no dataset:", len(tiktok))
```

```python
# Removendo dados NaN, NA, etc
# Não existem dados nulos neste dataset

tiktok.dropna(inplace=True)
tiktok.reset_index(drop=True, inplace=True)

print("Número de registros no dataset após remover as linhas com dados faltantes:", len(tiktok))
```

```python
# Lista todas as colunas do dataset

print(tiktok.columns)
```

```python
# Encontra os registros duplicados e os exibe

tiktok.drop(['Unnamed: 0', 'genre'], axis=1, inplace = True) # Tira colunas desnecessárias

duplicates = tiktok[tiktok.duplicated(keep='last')]

print("Número de registros duplicados:", len(duplicates))
duplicates.head()
```

```python
# Remove os registros duplciados

tiktok.drop_duplicates(inplace=True)
tiktok.reset_index(drop=True, inplace=True)

print("Número de registros sem as duplicatas:", len(tiktok))
```

```python
# Cria a coluna "year" para facilitar agregações

tiktok['release_year'] = [int(item[:4]) for item in tiktok['release_date']]

tiktok.head()
```

```python
# Exclui todas as músicas com popularidade zero

tiktok.drop(tiktok[tiktok['popularity'] == 0].index, inplace = True)
```

```python
# Cria um novo dataset que será efetivamente usado para as análises apenas com colunas numéricas e o ID da música

columns = ['track_name', 'artist_name', 'album_id', 'artist_id', 'playlist_id', 'playlist_name', 'release_date']

numeric_columns = tiktok.drop(columns, axis=1)
numeric_columns.head()
```
